## 1. Plot the distribution of  monthly income i.e the population distribution.

```python
# Plots the distribution of population i.e monthly income
sns.distplot(MI)   
plt.title("Population Distribution") 
    
# Calculating the mean and Standard deviation of population.
SD_ppl=np.std(MI) 
Mean_ppl=np.mean(MI)

print("Standard Deviation of population: ", SD_ppl)
print("Mean of population: ",Mean_ppl)
```

## 2. Draw a sample of size 200 and plot the sample distribution and comment on weather it represents the population distribution.

```python
# Creating a sample of size 200
sample=df.MonthlyIncome.sample(200,random_state=1)

# Plotting the distribution of sample and population
sns.distplot(sample)
plt.title("Sample Distribution")
plt.show()
sns.distplot(MI)   
plt.title("Population Distribution")
plt.show()
# Calculating the mean and Standard deviation of sample.
SD_sap=np.std(sample)   
Mean_sap=np.mean(sample)

print("Standard Deviation of sample: ", SD_sap)
print("Mean of sample ",Mean_sap)
```

#### Observations:
1. The sample distribution has the same kind of skewness in distribution, as population.
2. The Standard deviation and mean for both population and sample are approximately same.

#### Inference:

From the above observations we can say that the sample correctly represents the population.
